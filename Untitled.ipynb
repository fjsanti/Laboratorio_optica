$S_{i}(S_{0})=\frac{S_{0}f}{S_{0}-f}$

$\Delta{S}=0.05$

$S_{0}= (70, 115)$

Dado $f = 36.16$


| $S_{0}$ | $S_{i_{p}}$ |$S_{i_{t}}$|
|---|---|--|
| 49.8 | 134.9 | 132.0
| 56.2 | 98.8 | 101.4
| 57.2 | 98.6 | 98.3
| 59.1 | 91.0 | 93.1
| 67.1 | 77.7 | 78.4
| 72.0 | 72.4 | 72.6
| 72.1 | 72.3 | 72.5
| 76.1 | 68.8 | 68.9
| 78.0 | 68.2 | 67.4
| 111.1 | 54.8| 53.6

```python
import matplotlib.pyplot as plt
import numpy as np

# Parte teorica
x = np.arange(48, 115, 1) # Decides el rango de So con los dos primeros valores y el 1
# no lo muevas, es la suavidad de la curva
f = 36.16 # Defines tu focal
y = (f*x)/(x-f) # Defines S_i

# Parte practica
s_o = (49.8, 56.2, 57.2, 59.1, 67.1, 72.0, 72.1, 76.1, 78.0, 111.1) # Valor punto a punto de So
s_i = (134.9, 98.8, 98.6, 91.0, 77.7 , 72.4, 72.3, 68.8, 68.2, 54.8) # S_i
e = [0.05, 0.05, 0.05, 0.05, 0.05, 0.05, 0.05, 0.05, 0.05, 0.05] # Y su error para cada uno

# Para mayor es manera sería una terna como (49.8, 134.9, 0.05)

plt.plot(x, y)

upperlimits = [True, False] * 5
lowerlimits = [False, True] * 5

plt.xlabel('$S_{0}$') # Etiqueta de las horizontales
plt.ylabel('$S_{i}$') # Etiqueta de las verticales
plt.title('$f_{1}, f_{2},  f_{3}$') # El titulo que quieras en latex

plt.errorbar(s_o, s_i, yerr=e, xuplims=True, label='xuplims=True', fmt='o')
plt.legend(['$ S_{o}(S_{i})= \\frac{fS_{o}}{f-S_{o}}$','Valores prácticos']) # Etiquetas en orden
plt.grid()
plt.show()
```
